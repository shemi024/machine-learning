```python
#categorical variable and One HOT encoding
import pandas as pd
df = pd.read_csv("/content/drive/MyDrive/homeprices.csv")
df
```

```python
#using pandas to create  dummy variables
dummies = pd.get_dummies(df.town)
dummies
```

```python
merged = pd.concat([df,dummies],axis='columns')
merged
```

```python
final = merged.drop(['town'],axis='columns')
final
```

```python
final = final.drop(['west windsor'], axis='columns')
final
```

```python
x = final.drop(['price'], axis='columns')
x
```

```python
y = final.price
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(x,y)
```

```python
model.predict(x)
```

```python
model.predict([[3400,0,0]]) #3400 sqr ft home in west windsor
```

```python
model.predict([[2800,0,1]])
```

```python
#usin sklearn OneHotEncode
#first step os to use label encoder to convert town names into number
```

```python
from sklearn.preprocessing import LabelEncoder
le= LabelEncoder()
dfle= df
dfle.town = le.fit_transform(dfle.town)
dfle
```

```python
x= dfle[['town','area']].values
x
```

```python
y= dfle.price.values
y
```

```python
#now use one hot encoder to create dummy varibale for each for each of the town
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer
ct= ColumnTransformer([('town', OneHotEncoder(),[0])], remainder= 'passthrough')
x=ct.fit_transform(x)
x
```

```python
x= x[:,1:]
x
```

```python
model.fit(x,y)
```

```python
model.predict([[0,1,3400]])
```

```python
model.predict([[1,0,2800]])
```
